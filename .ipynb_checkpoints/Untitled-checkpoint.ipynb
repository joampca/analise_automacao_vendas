#### Importação e visualização dos dados

```python
# Importando bibliotecas para manipulação de dados e envio de email.

import pandas as pd 
import win32com.client as win32

# Carregamento dos dados.
pd.set_option('display.float_format', lambda x: '%.2f' % x)

data_vendas = pd.read_excel('Vendas.xlsx')
display(data_vendas.head())
```

#### Criando novos DF para coletar informações detalhadas

```python
# Faturamento por loja.

faturamento = data_vendas[['ID Loja', 'Valor Final']].groupby('ID Loja').sum()
display(faturamento.reset_index())
```

```python
# Quantidade de Produtos vendidas por Loja.

quantidade = data_vendas[['ID Loja','Quantidade']].groupby('ID Loja').sum()
display(quantidade.reset_index())
```

```python
# Ticket Médio por Loja.

ticket_medio = (faturamento['Valor Final'] / quantidade['Quantidade']).to_frame()
ticket_medio = ticket_medio.rename(columns={0:'Ticket Médio'})
display(ticket_medio.reset_index())
```

#### Criando E-mail para envio de relatório

```python
# Corpo do email para envios utilizando Outlook.

outlook = win32.Dispatch('outlook.application')
mail = outlook.CreateItem(0) # Item no outlook.
mail.To = 'joampca@outlook.com' # Destinatário
mail.Subject = 'Relatório de vendas' # Assunto
mail.HTMLBody = f'''<p>Prezados..</p> 

<p>Segue o relatório de vendas</p>

<p>Faturamento</p>
{faturamento.to_html()}

<p>Quantidade Vendida</p>
{quantidade.to_html()}

<p>Ticket Médio dos produtos por Loja</p>
{ticket_medio.to_html()}

<p>Atenciosamente.</p>
''' # Corpo do email padrão HTML

mail.Send()
```
